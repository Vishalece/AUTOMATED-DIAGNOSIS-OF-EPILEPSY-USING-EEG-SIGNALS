```python
import numpy as nm  
import matplotlib.pyplot as mtp  
import pandas as pd  
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearnex import patch_sklearn
patch_sklearn()

data_set= pd.read_csv('final_news.csv')  

x = data_set.iloc[:, 0:4].values
y= data_set.iloc[:, -1].values  

from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0)  
#feature Scaling  
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test)    

pd.DataFrame(x_train)
pd.DataFrame(x_test)
pd.DataFrame(y_train)
pd.DataFrame(y_test)
```

```python
modelall=[]
modelall.append(SVC())
modelall.append(GaussianNB())
modelall.append(LogisticRegression(random_state=0))
modelall.append(BernoulliNB())
modelall.append(KNeighborsClassifier())
modelall.append(DecisionTreeClassifier())
modelall.append(RandomForestClassifier())
```

```python
import time
start = time.time()
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearnex import patch_sklearn
patch_sklearn()

models= ['svm','GaussianNB','logistic','BernoulliNB','KNN','DECISION TREE','RANDOM']

acc=[]
cm=[]
for model in modelall:
    model.fit(x_train,y_train)
    y_pred= model.predict(x_test)
    acc.append(metrics.accuracy_score(y_test, y_pred)*100)
    cm.append(confusion_matrix(y_test,y_pred))
end = time.time()
print("Execution Time :",end-start,"sec")
print(acc)
```

```python
from sklearnex import patch_sklearn
patch_sklearn()
from xgboost.sklearn import XGBRFClassifier
from xgboost import XGBClassifier  

model = XGBRFClassifier()
model.fit(x_train,y_train)
pickle.dump(model,open('xgboost.pkl','wb'))
print(model)  
```

```python
y_pred= model.predict(x_test)  
accuracy = accuracy_score(y_test,y_pred)* 100.0
print("Accuracy: %.2f%%" % (accuracy ))  

acc.append(accuracy)

acc
```

```python
algo = ['SVM','GaussianNB','LogisticRegression','BernoulliNB','KNeighborsClassifier','DecisionTreeClassifier','RandomForestClassifier','XGBRFClassifier']
fig = plt.figure(figsize = (15, 5))
plt.bar(algo, acc, color ='skyblue',
        width = 0.4)
plt.xlabel("Classification Algorithms")
plt.ylabel("Accuracy of the model")
plt.title("Acuuracy of different classification algorithms ")
plt.show()
```

```python
fig, ax = plt.subplots(figsize =(20, 9))
ax.barh(algo, acc)
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)

ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)

ax.invert_yaxis()

for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str((i.get_width()).round(2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
plt.xlabel("Classification Algorithms")
plt.ylabel("Accuracy of the model")
plt.title("Acuuracy of different classification algorithms ")
plt.show()
```
